```python
import pandas as pd
from difflib import get_close_matches

# 1) If you've already read the parquet to `file`, keep using that:
# file = pd.read_parquet("<path>")

columns = [
    "dibjet_mass", "diphoton_mass", "bbgg_mass",
    "dibjet_pt", "diphoton_pt", "bbgg_pt",
    "lead_pho_pt", "sublead_pho_pt",
    "bbgg_eta", "bbgg_phi",
    "lead_pho_eta", "lead_pho_phi",
    "sublead_pho_eta", "sublead_pho_phi",
    "diphoton_eta", "diphoton_phi",
    "dibjet_eta", "dibjet_phi",
    "lead_bjet_pt", "sublead_bjet_pt",
    "lead_bjet_eta", "lead_bjet_phi",
    "sublead_bjet_eta", "sublead_bjet_phi",
    "sublead_bjet_PNetB", "lead_bjet_PNetB",
    "CosThetaStar_gg", "CosThetaStar_jj", "CosThetaStar_CS",
    "DeltaR_jg_min",
    "pholead_PtOverM", "phosublead_PtOverM",
    "FirstJet_PtOverM", "SecondJet_PtOverM",
    "lead_pt_over_diphoton_mass", "sublead_pt_over_diphoton_mass",
    "lead_pt_over_dibjet_mass", "sublead_pt_over_dibjet_mass",
    "diphoton_bbgg_mass", "dibjet_bbgg_mass",
    # the following names likely differ in your file:
    "lead_pho_mvaID_WP90", "lead_pho_mvaID_WP80",
    "sublead_pho_mvaID_WP90", "sublead_pho_mvaID_WP80",
    "lead_pho_mvaID", "sublead_pho_mvaID",
    "max_gamma_MVA_ID",
    "DeltaPhi_j1MET", "DeltaPhi_j2MET",
    # add more if you need them
]

# 2) Apply a simple rename map for obvious column name differences
rename_map = {
    "lead_pho_mvaID_WP90": "lead_mvaID_WP90",
    "lead_pho_mvaID_WP80": "lead_mvaID_WP80",
    "sublead_pho_mvaID_WP90": "sublead_mvaID_WP90",
    "sublead_pho_mvaID_WP80": "sublead_mvaID_WP80",
    "lead_pho_mvaID": "lead_mvaID",
    "sublead_pho_mvaID": "sublead_mvaID",
}

requested = [rename_map.get(c, c) for c in columns]

# 3) Keep only columns that exist; report missing + close matches
available = set(file.columns)
valid = [c for c in requested if c in available]
missing = [c for c in requested if c not in available]

print(f"Selecting {len(valid)} columns.")
if missing:
    print(f"{len(missing)} requested columns are missing. Showing close matches:")
    for m in missing:
        suggestions = get_close_matches(m, file.columns.tolist(), n=3, cutoff=0.6)
        print(f"  - {m}  ->  suggestions: {suggestions}")

# 4) Subset safely
df = file.loc[:, valid].copy()
```
