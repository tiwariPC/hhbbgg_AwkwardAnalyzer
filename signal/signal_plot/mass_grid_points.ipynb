```python
# %% [markdown]
# # Grid of (mX, mY) Scan Points
# 
# This cell creates a publication-quality scatter plot of (mX, mY) mass points.
# You can adjust marker size, colors, and figure settings interactively.

# %%
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# --- Mass point dictionary ---
mass_points = {
    300: [90, 95, 100, 125, 150, 170],
    320: [90, 95, 100, 125, 150, 170],
    350: [90, 95, 100, 125, 150, 170, 200],
    400: [90, 95, 100, 125, 150, 170, 200],
    450: [90, 95, 100, 125, 150, 170, 200, 250, 300],
    500: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350],
    550: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400],
    600: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450],
    650: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500],
    700: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550],
    750: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600],
    800: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650],
    850: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700],
    900: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700],
    950: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 800],
    1000: [90, 95, 100, 125, 150, 170, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 800],
}

# --- Plot parameters ---
cmap_name = "tab20"    # Try "tab10" or "viridis" for different looks
marker_size = 110
figsize = (11, 6.5)

# --- Prepare flattened data arrays ---
xs, ys = [], []
for mx, my_list in mass_points.items():
    for my in my_list:
        xs.append(mx)
        ys.append(my)
xs = np.array(xs)
ys = np.array(ys)

unique_y = sorted(set(ys))
y_to_idx = {y: i for i, y in enumerate(unique_y)}
color_indices = np.array([y_to_idx[y] for y in ys])

# --- Create plot ---
fig, ax = plt.subplots(figsize=figsize, dpi=120)
cmap = plt.get_cmap(cmap_name, len(unique_y))

sc = ax.scatter(xs, ys, c=color_indices, cmap=cmap, s=marker_size,
                edgecolors='black', linewidths=0.6, alpha=0.95, zorder=3)

# --- Labels and titles ---
ax.set_xlabel(r'$m_X$ (Spin-0) [GeV]', fontsize=14, labelpad=8)
ax.set_ylabel(r'$m_Y$ [GeV]', fontsize=14, labelpad=8)
ax.set_title('Grid of ($m_X$, $m_Y$) Scan Points', fontsize=16, pad=10)

# --- Axis ticks ---
mx_vals = sorted(mass_points.keys())
ax.set_xticks(mx_vals)
ax.set_xticklabels(mx_vals, fontsize=11)
ax.set_yticks(sorted(unique_y))
ax.set_yticklabels(sorted(unique_y), fontsize=10)

# --- Grid and frame style ---
ax.grid(which='major', linestyle='--', linewidth=0.6, alpha=0.5, zorder=0)
for spine in ax.spines.values():
    spine.set_linewidth(0.8)

# --- Legend creation ---
legend_handles = [
    Line2D([0], [0], marker='o', color='w',
           markerfacecolor=cmap(i), markeredgecolor='black',
           markersize=8, label=f"{y} GeV")
    for i, y in enumerate(sorted(unique_y))
]
ncol = 2 if len(legend_handles) > 12 else 1
ax.legend(
    handles=legend_handles,
    title=r"$m_Y$ [GeV]",
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    fontsize=10,
    title_fontsize=11,
    frameon=True,
    ncol=ncol,
    borderpad=0.8,
    labelspacing=0.6,
)

plt.subplots_adjust(right=0.78, top=0.9, left=0.08)
plt.show()
```
