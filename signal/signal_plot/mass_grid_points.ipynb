```python
import matplotlib.pyplot as plt

# Define the mass points you provided
mass_points = {
    300: [60, 70, 80, 90, 95, 100, 125, 150],
    400: [60, 70, 80, 90, 95, 100, 125, 150, 200],
    500: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300],
    550: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400],
    600: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400],
    650: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400, 500],
    700: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400, 500],
    800: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400, 500, 600],
    900: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400, 500, 600],
    1000: [60, 70, 80, 90, 95, 100, 125, 150, 200, 300, 400, 500, 600, 800]
}

# Setup the plot
plt.figure(figsize=(10, 6))

colors = plt.cm.tab20.colors  # Use a colormap with many distinct colors
color_index = 0

# Flatten unique Y values to assign a color to each
unique_y_values = sorted({y for ys in mass_points.values() for y in ys})
y_color_map = {y: colors[i % len(colors)] for i, y in enumerate(unique_y_values)}

# Plot each point
for mX, mYs in mass_points.items():
    for mY in mYs:
        plt.scatter(mX, mY, color=y_color_map[mY], s=40)

# Labels and grid
plt.xlabel(r'$m_X$ (Spin-0) [GeV]', fontsize=14)
plt.ylabel(r'$m_Y$ [GeV]', fontsize=14)
plt.grid(True, linestyle=':', color='black', alpha=0.2)

# Create a legend
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=y_color_map[y], markersize=8, label=str(y)) for y in unique_y_values]
plt.legend(handles=handles, title='Y', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

plt.tight_layout()
plt.title('X-Y Mass Point Grid')
plt.savefig("/afs/cern.ch/user/s/sraj/sraj/www/CUA/HH-bbgg/signal_study/grid_point_mx_1000.png")
plt.savefig("/afs/cern.ch/user/s/sraj/sraj/www/CUA/HH-bbgg/signal_study/grid_point_mx_1000.pdf")
plt.show()
```
